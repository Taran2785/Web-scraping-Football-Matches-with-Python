# Web Scraping Football Matches with Python

```python
#import Libraries
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#Access web URL
web = rq.get("https://fbref.com/en/comps/9/Premier-League-Stats")
soup= bs(web.text)

#Extract match data stats
standings_table= soup.select("table",{'class':'table.stats_table'})[0]
links = standings_table.find_all("a")
links= [l.get("href")for l in links]
links=[l for l in links if '/squads/' in l]
links
```

```python
#convert team urls to http
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

```python
#convert team_urls to dataframe
team_url = team_urls[0]
data=rq.get(team_url)
matches= pd.read_html(data.text,match='Scores & Fixtures')
matches
```

```python
#Get match shooting stats

soup= bs(data.text)
links= soup.find_all("a")
links= [l.get("href")for l in links]
links=[l for l in links if l and 'all_comps/shooting/' in l]
links
```

```python
#convert links to dataframe
data = rq.get(f"https://fbref.com{links[0]}")
shooting=  pd.read_html(data.text,match="Shooting")[0]
shooting
```
